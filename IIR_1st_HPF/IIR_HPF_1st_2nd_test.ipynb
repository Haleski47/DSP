```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqz
fs = 10000                  # 采样频率
fc_fs = 2e-6                # fc/fs (归一化频率)
fc = fc_fs * fs 
b,a = butter(1, 2 * fc_fs, btype="high")
scale = 1 / 2**26
# b = np.round(b / scale) * scale
# a = np.round(a / scale) * scale
print(b,a)
# 生成低频分析频率点（Hz）
f_low = np.linspace(0, 0.01, 10000)  # 分析 0 ~ 1 Hz
w = 2 * np.pi * f_low / fs       # 转换为 rad/sample

# 频率响应
w, h_1 = freqz(b, a, worN=w)

# 绘图
plt.plot(f_low, 20 * np.log10(np.abs(h_1)))
plt.title(f'High-pass Filter Response (fc={fc:.6f} Hz, fs={fs} Hz)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude (dB)')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.signal import butter, freqz

fs = 10000                  # 采样频率
fc_fs = 2e-6               # fc/fs (归一化频率)
fc = fc_fs * fs
b1, a1 = butter(1, 2 * fc_fs, btype="high")
scale = 1 / 2**26
# b1 = np.round(b1 / scale) * scale
# a1 = np.round(a1 / scale) * scale
print("1阶滤波器系数:", b1, a1)

# 分析频率点
f_low = np.linspace(0, 5, 100000)  # 0 ~ 5 Hz
w = 2 * np.pi * f_low / fs        # 转换为 rad/sample

# 一阶滤波器频率响应
w1, h_1 = freqz(b1, a1, worN=w)

# 相位延迟计算（Phase Delay）
phase_deg_1 = np.angle(h_1, deg=True)

# 二阶滤波器
fc_fs = 2e-6
fc = fc_fs * fs
b2, a2 = butter(2, 2 * fc_fs, btype="high")
# b2 = np.round(b2 / scale) * scale
# a2 = np.round(a2 / scale) * scale
print("2阶滤波器系数:", b2, a2)

# 二阶频率响应
w2, h_2 = freqz(b2, a2, worN=w)
phase_deg_2 = np.angle(h_2, deg=True)


# ========== 绘图部分 ==========
plt.figure(figsize=(12, 6))
plt.subplot(2, 1, 1)
plt.plot(f_low, 20 * np.log10(np.abs(h_1)), label='1st Order IIR HPF')
plt.plot(f_low, 20 * np.log10(np.abs(h_2)), label='2nd Order IIR HPF')
plt.title(f'Magnitude Response (fc={fc:.6f} Hz, fs={fs} Hz)')
plt.ylabel('Magnitude (dB)')
plt.grid(True)
plt.legend()

plt.subplot(2, 1, 2)
plt.plot(f_low, phase_deg_1, label='Phase Response - 1st Order')
plt.plot(f_low, phase_deg_2, label='Phase Response - 2nd Order')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Phase (degrees)')
plt.title('Phase Response of IIR High-pass Filters')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
